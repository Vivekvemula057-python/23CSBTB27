```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

```python
dataset = pd.read_csv("/content/breast-cancer.csv")
dataset.head(10)
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
num_columns = []

for feature in dataset.columns:
    if dataset[feature].dtypes != 'O':
        num_columns.append(feature)

print(num_columns)
```

```python
print(len(num_columns))
```

```python
Label = LabelEncoder()
dataset['diagnosis'] = Label.fit_transform(dataset['diagnosis'])
dataset['diagnosis'].head()
```

```python
target = dataset['diagnosis']
target.head()
```

```python
plt.figure(figsize = (10,10))
```

```python
dataset.groupby('id')['diagnosis'].plot.bar()
```

```python
dataset.drop(columns = 'id', inplace =True)
num_columns.remove('id')
dataset.head()
```

```python
discrete_feature = []
continuous_feature = []
for feature in num_columns:
    if dataset[feature].nunique() < 15:
        discrete_feature.append(feature)
    else:
        continuous_feature.append(feature)
print(discrete_feature)
print(continuous_feature)
```

```python
len(continuous_feature)
```

```python
dataset.shape
```

```python
dataset.groupby('diagnosis').count()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
data = dataset.copy()
for feature in continuous_feature:
    sns.histplot(data[feature], kde = True)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(dataset.corr(),annot = True , fmt = '0.2f', square = True)
```

```python
data = dataset.copy()

for feature in continuous_feature:
    plt.boxplot(data[feature])
    plt.xlabel(feature)
    plt.show()
```
